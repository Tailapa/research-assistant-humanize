```python
from crewai import Agent
from crewai_tools import DuckDuckGoSearchTool

# No API key setup needed!
ddg_search = DuckDuckGoSearchTool()

evidence_collector = Agent(
    role="Lead Fact-Checker",
    goal="Gather empirical data points from the open web",
    backstory="An expert investigator skilled at parsing search engine results.",
    llm=llm,
    tools=[ddg_search],
    verbose=True
)
```

#### 2. Serper.dev (Google Search API - 2,500 Free Requests)
1.  If your agents need high-quality, precise Google Search results, `Serper.dev` is the preferred industry standard for agentic frameworks.

2.  **The Free Plan**: Upon creating an account, they credit you with **2,500 free search requests**.

3.  **Why it's great**: Unlike scraping solutions, it returns structured `JSON` payloads containing site descriptions, organic rankings, answer boxes, and site links, which <u>Gemini can parse with exceptional accuracy.</u>

4.  **How to implement it**:
    *   Sign up at `serper.dev` to get your credential token, then configure it:

```python
import os
from crewai import Agent
from crewai_tools import SerperDevTool

# Set the environment token
os.environ["SERPER_API_KEY"] = "your_free_serper_key_here"

# Initialize the tool
google_search = SerperDevTool()

# Add to your agent's tool list
researcher = Agent(
    role="Data Acquisition Specialist",
    goal="Query Google to discover raw evidence matching target research criteria.",
    llm=llm,
    tools=[google_search],
    verbose=True
)
```

#### 3. Tavily AI (Built specifically for LLM Agents - 1,000 Free Requests/Month)
1.  Tavily is a search engine built from the ground up **for AI agents and LLMs**. Rather than returning raw web links that your agent has to manually navigate, Tavily searches the web, extracts the core textual facts, scrubs out irrelevant HTML clutter, and sends a clean, factual summary directly into your agent's prompt window.

2.  The Free Plan: Generously offers **1,000 search queries per month forever**.

3.  **How to implement it:**
    *   Sign up at `tavily.com`, grab an API key, and set it up using CrewAI's integration:

```python
import os
from crewai import Agent
from crewai_tools import TavilySearchTool

# Configure environment credentials
os.environ["TAVILY_API_KEY"] = "your_free_tavily_key_here"

# Initialize the tailored agent search tool
agent_search = TavilySearchTool()

evidence_collector = Agent(
    role="Synthesized Intelligence Researcher",
    goal="Extract direct factual summaries from modern online documentation.",
    llm=llm,
    tools=[agent_search],
    verbose=True
)
```
